```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use ("dark_background")
```

```python
dataset = pd.read_csv('AppleStore.csv')
dataset
```

```python
dataset.info()
```

```python
 dataset.isnull().sum()
```

```python
dataset.nunique()
```

```python
dataset.drop(['track_name'],axis = 1,inplace=True)
dataset.drop(['ver'],axis = 1,inplace=True)
dataset.drop(['Unnamed: 0'],axis = 1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
LE = LabelEncoder()
dataset['currency'] = LE.fit_transform(dataset['currency'])
dataset['cont_rating'] = LE.fit_transform(dataset['cont_rating'])
dataset['prime_genre'] = LE.fit_transform(dataset['prime_genre'])
```

```python
data =pd.DataFrame(dataset)
data
```

```python
data.info()
```

```python
sns.heatmap(data)
sns.pairplot (data)
```

```python
x = data[['id','size_bytes','currency','rating_count_ver','user_rating','user_rating_ver','cont_rating','prime_genre','sup_devices.num','ipadSc_urls.num','lang.num','vpp_lic']].values
x
```

```python
y = data[['price','rating_count_tot']].values
y
```

```python
SC = StandardScaler()
x[:,0:2] = SC.fit_transform(x[:,0:2])
x[:,3:-1] = SC.fit_transform(x[:,3:-1])
```

```python
df_x = pd.DataFrame(x , columns=['id','size_bytes','currency','rating_count_ver','user_rating','user_rating_ver','cont_rating','prime_genre','sup_devices.num','ipadSc_urls.num','lang.num','vpp_lic'])
df_x
```

```python
df_y = pd.DataFrame (y , columns=['price','rating_count_tot'])
df_y
```

```python
sns.heatmap(df_x)
```

```python
sns.heatmap(df_y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split (x ,y ,test_size = 0.2 ,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor (n_estimators = 1000,random_state = 0)
RFR.fit(x_train, y_train)
y_pred = RFR.predict(x_test)
y_pred
```
