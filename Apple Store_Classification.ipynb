```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use ("dark_background")
```

```python
dataset = pd.read_csv('AppleStore.csv')
dataset
```

```python
dataset.info()
dataset.nunique()
```

```python
dataset.drop(['track_name'],axis = 1,inplace=True)
dataset.drop(['ver'],axis = 1,inplace=True)
dataset.drop(['Unnamed: 0'],axis = 1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
LE = LabelEncoder()
dataset['currency']    = LE.fit_transform(dataset['currency'])
dataset['cont_rating'] = LE.fit_transform(dataset['cont_rating'])
dataset['prime_genre'] = LE.fit_transform(dataset['prime_genre'])
```

```python
sns.heatmap  (dataset)
sns.pairplot (dataset)
```

```python
x = dataset.iloc [:,:-1].values
x
```

```python
y = dataset.iloc[:,-1].values
y
```

```python
SC = StandardScaler()
x[:,:2 ] = SC.fit_transform(x[:,:2 ])
x[:,3:8] = SC.fit_transform(x[:,3:8])
x[:,9: ] = SC.fit_transform(x[:,9: ])
```

```python
df_x = pd.DataFrame(x ,columns=['id','size_bytes','currency','price','rating_count_tot','rating_count_ver','user_rating','user_rating_ver','cont_rating','prime_genre','sup_devices.num','ipadSc_urls.num','lang.num'])
df_x
```

```python
df_y = pd.DataFrame(y ,columns=['vpp_lic'])
df_y
```

```python
sns.heatmap  (df_x)
sns.pairplot (df_x)
```

```python
sns.heatmap  (df_y)
sns.pairplot (df_y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split (x ,y ,test_size = 0.2 ,random_state = 0)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print(y_pred)
```

```python
df_y_test = pd.DataFrame(y_test ,columns=['vpp_lic'])
df_y_test
```

```python
df_y_pred = pd.DataFrame(y_pred ,columns=['vpp_lic'])
df_y_pred
```

```python
df_comp = pd.DataFrame({'Actual':y_test , 'Predict':y_pred})
df_comp
```

```python
df_y_test.sum()
```

```python
df_y_pred.sum()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
